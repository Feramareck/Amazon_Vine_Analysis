```python
import pandas as pd
```

```python
# Create a DataFrame
vine_csv = "./Resources/vine_table.csv"

# Read the data
vine_df = pd.read_csv(vine_csv)
vine_df.head()
```

```python
# retrieve all the rows where the total_votes count is equalto or greater than 20
vine_df_total_votes = vine_df[vine_df["total_votes"] >= 20]
vine_df_total_votes.head()
```

```python
vine_df_total_votes.info()
```

```python
# retrieve all the rowswhere the number of helpful_votes divided by total_votes is equal to or greater than 50%
vine_df_helpful = vine_df_total_votes[(vine_df_total_votes["helpful_votes"]/vine_df_total_votes["total_votes"]) >= 0.5]
vine_df_helpful.head()
```

```python
# retrieve all the rows where a review was written as part of the Vine program (paid)
vine_df_program = vine_df_helpful[vine_df_helpful["vine"] == 'Y']
vine_df_program.head()
```

```python
# retrieve all the rows where the review was not part of the Vine program (unpaid)
vine_df_program_N = vine_df_helpful[vine_df_helpful["vine"] == 'N']
vine_df_program_N.head()
```

```python
#total number of reviews
vine_total = vine_df_helpful[vine_df_helpful.columns[0]].count()
# number of 5-star reviews
five_stars = vine_df_helpful[vine_df_helpful["star_rating"] == 5]
```

```python
# percentage of 5-star reviews (paid)
five_stars_Y_ = vine_df_program[vine_df_program["star_rating"] == 5]
five_stars_Y = ((five_stars_Y_[five_stars_Y_.columns[0]].count())/vine_total)*100
```

```python
# percentage of 5-star reviews (unpaid)
five_stars_N_ = vine_df_program_N[vine_df_program_N["star_rating"] == 5]
five_stars_N = ((five_stars_N_[five_stars_N_.columns[0]].count())/vine_total)*100
```

```python
#Summary dataframe
vine_anlysis_review = pd.DataFrame({'Total_Reviews': [vine_total],
                      '5_stars': [five_stars[five_stars.columns[0]].count()],
                      'Vine_Y':  [vine_df_program[vine_df_program.columns[0]].count()],
                      'Vine_N':  [vine_df_program_N[vine_df_program_N.columns[0]].count()], 
                      'Vine_Y_5_Stars':  [(five_stars_Y_[five_stars_Y_.columns[0]].count())],
                      'Vine_N_5_Stars':  [(five_stars_N_[five_stars_N_.columns[0]].count())],             
                      'Perc_5_stars_vine_Y':  [five_stars_Y],
                      'Perc_5_stars_vine_N': [five_stars_N]
                     })
vine_anlysis_review
```
